There are missing data across the rows and columns

```python
# To remove the the rows with missing data:

covid_dropped = covid_maindata.dropna()
covid_dropped
```

```python
# rename variable

covid = covid_dropped 
```

### To round up the values for latitude and longitude

To round up "Lat" to 2 decimal place:

```python
covid["Lat"] = covid["Lat"].round (decimals = 2)
```

```python
covid
```

To round up "Long" to 2 decimal place:

```python
covid ["Long"] = covid ["Long"].round (decimals = 2)
```

```python
covid
```

Rename Lat and Long

```python
covid_Latrename = covid.rename (columns = {"Lat" : "Latitude"})
```

```python
covid_Latrename
```

```python
covid_Longrename = covid_Latrename.rename (columns = {"Long" : "Longitude"})
```

```python
covid_Longrename
```

```python
#rename variable

covid_data = covid_Longrename
```

Select "United Kingdom" Data

```python
covid_uk = covid_data [covid_data["Country/Region"] == "United Kingdom"]
covid_uk
```

Rename United Kingdom to UK

```python
covid_UK = covid_uk.replace ("United Kingdom", "UK")
covid_UK
```

```python
covid_UK.info
```

```python
covid_UK.dtypes
```

### PLOT DATA

```python
covid_UK.plot("Province/State", "5/17/22", kind = "bar");
```

```python
# setup plot
fig, ax = plt.subplots (nrows = 2, ncols = 2, figsize = (15, 10))

# plot data on each axis
ax[0,0].barh(covid_UK["Province/State"], covid_UK["1/10/22"].astype(int));
ax[0,1].barh(covid_UK["Province/State"], covid_UK["2/20/22"].astype(int));
ax[1,0].barh(covid_UK["Province/State"], covid_UK["3/30/22"].astype(int));
ax[1,1].barh(covid_UK["Province/State"], covid_UK["4/26/22"].astype(int));

# add title to each subplot
ax[0, 0].set_title ("1/10/2022");
ax[0, 1].set_title ("2/20/2022");
ax[1, 0].set_title ("3/30/2022");
ax[1, 1].set_title ("4/26/2022");

# add main title to figure
plt.suptitle ("Covid cases realtime data analysis for UK Provinces: January 2022-April 2022".upper(), fontsize = 20);

# add space between subplots
fig.tight_layout(pad = 5.5)

```
